```python
import pandas as pd
import re
import csv
```

```python
text = "Baru masuk aplikasi mau aktifin mbanking kode akses udah di isi -Persetujuan Selanjutnya menunggu Indikator warna hijau ,INI INDIKATOR NYA AJA GA ADA GIMANA MAU TAU IJO SAMA BIRU UDAH 6 KALI NYOBA GAGAL TERUS ‼️"
```

CASE FOLDING

```python
def text_cleaning(text):
    text = text.lower()                           # case folding
    text = re.sub(r'http\S+|www\S+', '', text)   # hapus URL
    text = re.sub(r'\d+', '', text)              # hapus angka
    text = re.sub(r'[^\w\s]', '', text)          # hapus simbol/tanda baca/emoji
    text = re.sub(r'\s+', ' ', text).strip()     # rapihin spasi
    
    return text

cleaned_text = text_cleaning(text)

print(cleaned_text)
```

Normalisasi Kata

```python
slang_dict = {
    "aktifin": "aktifkan",
    "udah": "sudah",
    "ga": "tidak",
    "tau": "tahu",
    "ijo": "hijau",
    "nyoba": "mencoba",
    "mbanking": "mobile banking",
    "nya": ""
}

def normalize_text(text):
    words = text.split()

    normalized = [
        slang_dict[word] if word in slang_dict else word
        for word in words
    ]

    return " ".join(normalized)

normalized_text = normalize_text(cleaned_text)

print(normalized_text)
```

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory

factory = StopWordRemoverFactory()
stopwords = factory.get_stop_words()

def remove_stopwords(text):
    words = text.split()

    filtered_words = [
        word for word in words
        if word not in stopwords
    ]

    return " ".join(filtered_words)

final_text = remove_stopwords(normalized_text)

print(final_text)
```
